```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12,10)
```

```python
with open('r50res_32batches(size64)_sid0.pkl','rb') as f:
    res = pickle.load(f)
```

```python
res.keys()
```

```python
clado_perf = [res['clado_perf'][i]['top1'].item() for i in range(len(res['clado_perf']))]
```

```python
naive_perf = [res['naive_perf'][i]['top1'].item() for i in range(len(res['naive_perf']))]
mpqco_perf = [res['mpqco_perf'][i]['top1'].item() for i in range(len(res['mpqco_perf']))]
```

```python
plt.rcParams['figure.figsize'] = (12,10)
plt.plot(res['clado_size'],clado_perf,marker='o',label='clado')
plt.plot(res['naive_size'],naive_perf,marker='o',label='naive')
plt.plot(res['mpqco_size'],mpqco_perf,marker='o',label='mpqco')
plt.legend()
plt.xlabel('model size(MB)')
plt.ylabel('test accuracy')
```

```python
plt.rcParams['figure.figsize'] = (12,10)
plt.plot(res['clado_bitops'],clado_perf,marker='o',label='clado')
plt.plot(res['naive_bitops'],naive_perf,marker='o',label='naive')
plt.plot(res['mpqco_bitops'],mpqco_perf,marker='o',label='mpqco')
plt.legend()
plt.xlabel('bitoperations (G)')
plt.ylabel('test accuracy')
```
