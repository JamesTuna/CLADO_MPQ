```python
int4params,_ = pyhessian.get_params_grad(int4model)
bfp12params,_ = pyhessian.get_params_grad(bfp12model)
fp32params,_ = pyhessian.get_params_grad(model)
int4_error = [(x-y) for x,y in zip(int4params,fp32params)]
bfp12_error = [(x-y) for x,y in zip(bfp12params,fp32params)]
```

```python
int4_l2 = pyhessian.group_product(int4_error,int4_error)**0.5
bfp12_l2 = pyhessian.group_product(bfp12_error,bfp12_error)**0.5
```

```python
int4_error_var = [x.std() for x in int4_error]
bfp12_error_var = [x.std() for x in bfp12_error]
```

```python
import matplotlib.pyplot as plt
import numpy as np

for x,y,int4err,bfp12err in zip(int4_error_var,bfp12_error_var,int4_error,bfp12_error):
    if x == 0:
        continue
    print('tensor shape',int4err.shape)
    print('normalized statistics (unit l2 error)')
    print('std of int 4 error',x/int4_l2,'\nstd of bfp12 error',y/bfp12_l2)
    print('raw statistics')
    print('std of int 4 error',x,'\nstd of bfp12 error',y)
    print()
#     plt.clf()
#     plt.hist(int4err.cpu().detach().numpy().flatten(),bins=30,alpha=0.5,label='int4 error')
#     plt.hist(bfp12err.cpu().detach().numpy().flatten(),bins=30,alpha=0.5,label='bfp12 error')
#     #plt.hist(int4err.cpu().detach().numpy().flatten()/int4_l2.cpu().detach().numpy(),alpha=0.5,label='int4 error')
#     #plt.hist(bfp12err.cpu().detach().numpy().flatten()/bfp12_l2.cpu().detach().numpy(),alpha=0.5,label='bfp12 error')
#     plt.legend()
#     plt.show()
    
    plt.clf()
    num_bins = 20
    normalized_int4err = int4err.cpu().detach().numpy()/int4_l2.cpu().detach().numpy()
    normalized_bfp12err = bfp12err.cpu().detach().numpy()/bfp12_l2.cpu().detach().numpy()
    
    normalized_int4err = int4err.cpu().detach().numpy()
    normalized_bfp12err = bfp12err.cpu().detach().numpy()
    allerr = np.concatenate([normalized_int4err,normalized_bfp12err])
    _,bins = np.histogram(allerr,num_bins)
    
    
    heights,bins = np.histogram(int4err.cpu().detach().numpy()/int4_l2.cpu().detach().numpy(),bins)
    freq = heights/sum(heights)
    width = (max(bins)-min(bins))/(len(bins))
    plt.bar(bins[:-1]+1/2*width,freq,width=width,color='blue',alpha=0.5,label='int4 error')
    
    heights,bins = np.histogram(bfp12err.cpu().detach().numpy()/bfp12_l2.cpu().detach().numpy(),bins)
    freq = heights/sum(heights)
    width = (max(bins)-min(bins))/(len(bins))
    plt.bar(bins[:-1]+1/2*width,freq,width=width,color='red',alpha=0.5,label='bfp12 error')
    
    plt.xticks(np.around(bins,decimals=2))
    plt.xlabel('Error')
    plt.ylabel('Frequency')
    plt.legend()
    plt.rc('grid', linestyle="--", color='black')
    plt.grid()
    plt.show()
```

```python
def replaceModuleByName(modelName,moduleName,newModuleName):
    '''
        replace module with name modelName.moduleName with newModule
    '''
    tokens = moduleName.split('.')
    eval_str = modelName
    for token in tokens:
        try:
            eval_str += f'[{int(token)}]'
        except:
            eval_str += f'.{token}'
            
    exec(eval_str+f'={newModuleName}')
    
for name,module in model.named_modules():
    if isinstance(module,torch.nn.Conv2d):
        #print(name,' is a conv2d')
        newLayer = qConv2d(0,0,0,init_from=module).cuda()
        replaceModuleByName('model',name,'newLayer')
    elif isinstance(module,torch.nn.Linear):
        #print(name,' is a linear')
        newLayer = qLinear(0,0,init_from=module).cuda()
        replaceModuleByName('model',name,'newLayer')
```

```python
# get all params that have variation on
def get_params_grad(model):
    """
    get model parameters and corresponding gradients
    """
    params = []
    grads = []
    for name,module in model.named_modules():
        if not isinstance(module,qConv2d) and not isinstance(module,qLinear):
            continue
        param = module.weight
        params.append(param)
        grads.append(0. if param.grad is None else param.grad + 0.)
    return params, grads
```

```python
import tensorboardX
def QAVATPLUS_train(model,train_loader,test_loader,config,imgSize=32,imgFlat=False,
                lossfunc=torch.nn.CrossEntropyLoss(),printPerEpoch=100):

    tb = tensorboardX.SummaryWriter(comment=config['trial_name'])
    C = config
    
    #params, grad = pyhessian.get_params_grad(model)
    
    if C['optimizer'] == 'SGD' or C['optimizer'] == 'sgd':
        optimizer = torch.optim.SGD(model.parameters(),lr=C['lr'],momentum=0.9)
    elif C['optimizer'] == 'adam' or C['optimizer'] == 'Adam' or C['optimizer'] == 'ADAM':
        optimizer = torch.optim.Adam(model.parameters(),lr=C['lr'])
    else:
        print('unrecognized optimizer defined in config')
        exit(0)
    for epoch in range(C['epochs']):
        # lr decay
        current_lr = C['lr'] * (C['decay_ratio'] ** (epoch // C['decay_ep']))
        for param_group in optimizer.param_groups:
            param_group['lr'] = current_lr

        start = time.time()
        total_loss = 0
        batch_count = 0
        # per epoch training, do
        for i, data in enumerate(train_loader, 0):
            model.train()
            x, label = data
            if imgFlat:
                x = x.view(-1,imgSize**2)
            x = x.to(C['device'])
            label = label.to(C['device'])
            optimizer.zero_grad()
            '''
            if config['noise_std'] > 0:
                generate_variation(model,noise_std=config['noise_std'])
            output = model(x)
            l = lossfunc(output,label)
            l.backward()
            '''
            
            # use under-variation 2nd-order gradient estimation
            # g + sigma^2 (v*H(v*w))
            # v: rademacher variable
            # *: elementwise product
            # note that (v*H(v*w))i = Hii wi
            # Hv is the hessian-parameter product
            # computed using pyhessian (reuse gradient)
            
            output = model(x)
            l = lossfunc(output,label)
            l.backward(create_graph=True)
            params, grad = get_params_grad(model)
            
            v = [torch.randint_like(p, high=2, device='cuda') for p in params]
            for v_i in v:
                v_i[v_i == 0] = -1
            
            hv = torch.autograd.grad(grad,params,grad_outputs=v,retain_graph=False)
            hessianD = [torch.mul(v_i,hv_i) for v_i,hv_i in zip(v,hv)]
            
            #g_l2, e_g_l2 = 0,0
            
            for p,hessianD_ in zip(params,hessianD):
                extra_gradient = hessianD_ * p * (config['noise_std']**2)
                #g_l2 += torch.sum((p.grad)**2)
                #e_g_l2 += torch.sum(extra_gradient**2)
                p.grad.data += extra_gradient 
            
            #print('gradient magnitude',g_l2**0.5)
            #print('extra gradient magnitude',e_g_l2**0.5)
            optimizer.step()
            total_loss += l.data.item()
            batch_count += 1
        
        total_loss /= batch_count
        tb.add_scalar('epoch loss',total_loss,epoch+1)
        tb.add_scalar('epoch time',time.time()-start,epoch+1)
        tb.add_scalar('learning rate',current_lr,epoch+1)

        # console output
        if epoch % printPerEpoch == printPerEpoch-1:
            print("epoch %s loss %s [%.4f seconds]"%(epoch+1,total_loss,time.time()-start))

        if C['valPerEp'] is None:
            continue

        # validation
        if epoch % C['valPerEp'] == 0:
            val = evaluate(test_loader,model,noise_std = C['noise_std'],repeat = C['valSample'],imgSize=imgSize,imgFlat=imgFlat,device = C['device'])

            tb.add_scalar('validation/mean accuracy',val['mean_acc'],epoch+1)
            tb.add_scalar('validation/qtl accuracy',val['qtl_acc'],epoch+1)
            tb.add_scalar('validation/mean loss',val['mean_loss'],epoch+1)
            tb.add_scalar('validation/qtl loss',val['qtl_loss'],epoch+1)
            tb.add_scalar('validation/validation time',val['test time'],epoch+1)
            tb.add_histogram('validation/accuracy',val['acc_list'],epoch+1)
            tb.add_histogram('validation/loss',val['loss_list'],epoch+1)

            print("Epoch %s validation [%.4f seconds]"%(epoch+1,val['test time']))
            print("mean acc %.4f mean loss %.4f"%(val['mean_acc'],val['mean_loss']))

    tb.close()
```

```python
model.train()
C['trial_name'] = 'qavat_plus_train_invertSign'
QAVATPLUS_train(model,train,test,config=C,imgSize=32,imgFlat=False,
                lossfunc=torch.nn.CrossEntropyLoss(),printPerEpoch=1)
```
